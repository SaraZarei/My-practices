```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.pipeline import Pipeline
```

```python
df=pd.read_csv('E:\work\DataSets\\automobileEDA.csv')
```

```python
#develop a linear model with one feature
x1=df[['engine-size']]
y1=df['price']
lms=LinearRegression()
lms.fit(x1,y1)
print('intercept:',lms.intercept_,'coefficient:',lms.coef_)
y1hat=lms.predict(x1)
#visualize results
ax1=sns.distplot(y1,hist=False,color='r',label='y')
sns.distplot(y1hat,hist=False,color='b',label='yhat')
#R-Squared
print('R-Squared:',lms.score(x1,y1))
#MSE
print('MSE:',mean_squared_error(y1,y1hat))
```

```python
#develop a polynomial model with one features with pipeline
x3=df[['engine-size']]
y3=df['price']
input=[('scale',StandardScaler()),('polynomial',PolynomialFeatures(degree=5)),
       ('model',LinearRegression())]
pipe=Pipeline(input)
pipe.fit(x3,y3)
y3hat=pipe.predict(x3)
#visualize results
ax1=sns.distplot(y3,hist=False,color='r',label='y')
sns.distplot(y3hat,hist=False,color='b',label='yhat')
#R-Squared
print('R-Squared:',pipe.score(x3,y3))
#MSE
print('MSE:',mean_squared_error(y3,y3hat))
```

```python
#use residualplot to define which function is better, linear or non-linear
sns.residplot(x1,y1)
```
